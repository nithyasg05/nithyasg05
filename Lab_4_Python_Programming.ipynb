```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
# Question 20
virginica_sepal_width = iris[iris['species'] == 'virginica']['sepal_width'].min()
print(f"The minimum sepal width for virginica is: {virginica_sepal_width}")
```

```python
# Question 21
mean_petal_length = iris['petal_length'].mean()
filtered_iris = iris[iris['petal_length'] > mean_petal_length]

print(f"Number of observations where petal_length > mean petal_length: {len(filtered_iris)}")
```

```python
# Question 22
# Check the number of versicolor
versicolor_count = len(iris[iris['species'] == 'versicolor'])
print(f"Number of versicolor: {versicolor_count}")

# Check the mean sepal_length
mean_sepal_length = round(iris['sepal_length'].mean(), 3)
print(f"Mean sepal_length rounded to 3 decimal places: {mean_sepal_length}")

# Check the number of unique species
unique_species_count = iris['species'].nunique()
print(f"Number of unique species: {unique_species_count}")

# Check the number of virginica
virginica_count = len(iris[iris['species'] == 'virginica'])
print(f"Number of virginica: {virginica_count}")
```

```python
# Question 23
correlation_matrix = iris.corr(numeric_only=True)

print("Pearson Correlation Matrix:")
display(correlation_matrix)

# Check the correlation for each pair
sepal_length_sepal_width_corr = correlation_matrix.loc['sepal_length', 'sepal_width']
petal_length_sepal_length_corr = correlation_matrix.loc['petal_length', 'sepal_length']
sepal_length_petal_width_corr = correlation_matrix.loc['sepal_length', 'petal_width']
petal_length_petal_width_corr = correlation_matrix.loc['petal_length', 'petal_width']

print(f"\nCorrelation between sepal_length and sepal_width: {sepal_length_sepal_width_corr}")
print(f"Correlation between petal_length and sepal_length: {petal_length_sepal_length_corr}")
print(f"Correlation between sepal_length and petal_width: {sepal_length_petal_width_corr}")
print(f"Correlation between petal_length and petal_width: {petal_length_petal_width_corr}")
```

```python
# Question 24
# The mean value of petal_length of setosa is 1.46 (rounded)
setosa_petal_length_mean = round(iris[iris['species'] == 'setosa']['petal_length'].mean(), 2)
print(f"Mean petal_length of setosa: {setosa_petal_length_mean}")

# setosa has the largest average sepal_width
average_sepal_width_by_species = iris.groupby('species')['sepal_width'].mean()
print(f"\nAverage sepal_width by species:\n{average_sepal_width_by_species}")
setosa_largest_sepal_width = average_sepal_width_by_species['setosa'] == average_sepal_width_by_species.max()
print(f"setosa has the largest average sepal_width: {setosa_largest_sepal_width}")


# The standard deviation of sepal_length of setosa is 0.35 (rounded)
setosa_sepal_length_std = round(iris[iris['species'] == 'setosa']['sepal_length'].std(), 2)
print(f"\nStandard deviation of sepal_length of setosa: {setosa_sepal_length_std}")


# virginica has largest sepal_length
largest_sepal_length_by_species = iris.groupby('species')['sepal_length'].max()
print(f"\nLargest sepal_length by species:\n{largest_sepal_length_by_species}")
virginica_largest_sepal_length = largest_sepal_length_by_species['virginica'] == largest_sepal_length_by_species.max()
print(f"virginica has largest sepal_length: {virginica_largest_sepal_length}")
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
# Question 25
# Filter the titanic dataset for passengers in 2nd class who survived
survived_second_class = titanic[(titanic['pclass'] == 2) & (titanic['survived'] == 1)]

# Get the number of such passengers
num_survived_second_class = len(survived_second_class)

print(f"Number of passengers in 2nd class who survived: {num_survived_second_class}")
```

```python
# Question 26
# Filter the titanic dataset for passengers in 1st class who did not survive
first_class_not_survived = titanic[(titanic['pclass'] == 1) & (titanic['survived'] == 0)]

# Get the age range (minimum and maximum age) for this group
age_range_first_class_not_survived = (first_class_not_survived['age'].min(), first_class_not_survived['age'].max())

print(f"The age range in the 1st class who did not survive is: {age_range_first_class_not_survived}")
```

```python
# Question 27
# Filter for female passengers
female_passengers = titanic[titanic['sex'] == 'female']

# Filter for female passengers who survived
female_survivors = female_passengers[female_passengers['survived'] == 1]

# Calculate the percentage of female survivors
percentage_female_survivors = (len(female_survivors) / len(female_passengers)) * 100

# Round to the nearest whole percent
rounded_percentage = round(percentage_female_survivors)

print(f"The percentage of female passengers who survived is: {rounded_percentage}%")
```

```python
# Question 28
# Check for missing values in the specified columns
missing_sex = titanic['sex'].isnull().any()
missing_embarked = titanic['embarked'].isnull().any()
missing_deck = titanic['deck'].isnull().any()
missing_age = titanic['age'].isnull().any()

print(f"Missing values in 'sex': {missing_sex}")
print(f"Missing values in 'embarked': {missing_embarked}")
print(f"Missing values in 'deck': {missing_deck}")
print(f"Missing values in 'age': {missing_age}")
```

```python
# Question 29
# Calculate survival rate by passenger class
survival_rate_by_class = titanic.groupby('pclass')['survived'].mean()
print(f"Survival rate by passenger class:\n{survival_rate_by_class}")

# Calculate survival rate by embarkation point
survival_rate_by_embarked = titanic.groupby('embarked')['survived'].mean()
print(f"\nSurvival rate by embarkation point:\n{survival_rate_by_embarked}")

# Calculate correlation between survival and passenger class
correlation_survival_pclass = titanic['survived'].corr(titanic['pclass'])
print(f"\nCorrelation between survival and passenger class: {correlation_survival_pclass}")

# Calculate correlation between survival and fare
correlation_survival_fare = titanic['survived'].corr(titanic['fare'])
print(f"Correlation between survival and fare: {correlation_survival_fare}")
```
